```python
from scipy import sparse

eye = np.eye(4)

print("원래 표현:\n", eye)

csr_sparse = sparse.csr_matrix(eye)

print("csr 표현:\n", csr_sparse)

data = np.ones(4)
rows = np.arange(4)
cols = np.arange(4)

coo_sparse = sparse.coo_matrix((data, (rows, cols)))

print("coo 표현:\n", coo_sparse)
```

```python
%matplotlib inline
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris_dataset = load_iris()

print("iris dataset의 key:\n", iris_dataset.keys())
```

```python
from sklearn.model_selection import train_test_split

X = iris_dataset['data']
y = iris_dataset['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=.25)

print('train data 모양:\n', X_train.shape)
print('test data shape:\n', X_test.shape)
print('train target shape:\n', y_train.shape)
print('test target shape:\n', y_test.shape)
```

```python
import mglearn

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, marker='o', figsize=(15,15))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5,2.9, 1, .2]])
print('X_new.shape:', X_new.shape)
```

```python
predict = knn.predict(X_new)
print("predict value:", predict)
print("predict label:", iris_dataset['target_names'][predict])
```

```python
predict_value = knn.predict(X_test)
print("predict value:\n", predict_value)
print("predict score:", np.mean(y_test == predict_value))
```

```python
print("accuracy on test data set : {:.2f}".format(knn.score(X_test, y_test)))
```
